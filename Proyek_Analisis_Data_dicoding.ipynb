**Eploratory Data Analysis Business question 1**

Program ini dimulai dengan memverifikasi apakah data yang digunakan memiliki kolom yang diperlukan, yaitu `product_category_name`, `review_score`, dan `order_item_id`. Setelah itu, data dikelompokkan berdasarkan kategori produk (`product_category_name`), dengan menghitung jumlah pembelian (`purchase_count`) dan rata-rata rating (`average_rating`) per kategori. Kemudian, data diurutkan dua kali: pertama berdasarkan jumlah pembelian terbanyak, dan kedua berdasarkan rating tertinggi. Hasil kedua urutan ini digabungkan untuk mendapatkan produk yang muncul di kedua kategori tersebut. Selanjutnya, program menampilkan lima produk teratas yang memiliki kombinasi pembelian terbanyak dan rating tertinggi, serta memberikan statistik deskriptif untuk kolom `purchase_count` dan `average_rating`. Terakhir, data ditampilkan dengan format yang lebih mudah dibaca, di mana jumlah pembelian dipisahkan dengan ribuan dan rating ditampilkan dengan dua angka desimal.

```python
# Memeriksa apakah kolom yang diperlukan ada
required_columns = ["product_category_name", "review_score", "order_item_id"]
if not all(col in product_favorit_akhir.columns for col in required_columns):
    raise ValueError(f"Data harus memiliki kolom: {required_columns}")

# Mengelompokkan data berdasarkan kategori produk dan menghitung pembelian unik berdasarkan order_item_id
grouped_data = product_favorit_akhir.groupby("product_category_name").agg(
    purchase_count=("order_item_id", "sum"),  # Menghitung jumlah pembelian
    average_rating=("review_score", "mean")  # Menghitung rata-rata rating
).reset_index()

# Menyaring produk dengan pembelian terbanyak dan rating tertinggi
# Mengurutkan berdasarkan pembelian terbanyak
top_purchased_products = grouped_data.sort_values(by="purchase_count", ascending=False)

# Mengurutkan berdasarkan rating tertinggi
top_rated_products = grouped_data.sort_values(by="average_rating", ascending=False)

# Menampilkan produk yang ada di kedua kategori: pembelian terbanyak dan rating tertinggi
# Produk yang ada di dua urutan teratas (terbaik)
top_combined_products = pd.merge(top_purchased_products[['product_category_name', 'purchase_count']],
                                  top_rated_products[['product_category_name', 'average_rating']],
                                  on="product_category_name")

# Menampilkan 5 produk dengan pembelian terbanyak dan rating tertinggi
top_combined_products[['product_category_name', 'purchase_count', 'average_rating']].head()

# Menampilkan statistik deskriptif untuk kolom numerik
top_combined_products[['purchase_count', 'average_rating']].describe()

# Menampilkan data dengan format yang lebih baik
top_combined_products[['product_category_name', 'purchase_count', 'average_rating']].style.format({
    'purchase_count': '{:,}',  # Menampilkan jumlah pembelian dengan pemisah ribuan
    'average_rating': '{:.2f}'  # Menampilkan rating dengan dua angka desimal
})
```

Kemudian, data dikelompokkan berdasarkan kota, dan jumlah total pembelian dihitung untuk setiap kota dengan menggunakan `groupby` dan fungsi agregasi `count` pada kolom `order_item_id`. Setelah itu, data diurutkan berdasarkan jumlah pembelian terbanyak, dan 10 kota dengan pembelian tertinggi dipilih. Hasilnya, program menampilkan daftar 10 kota dengan jumlah pembelian terbanyak, di mana jumlah pembelian untuk setiap kota ditampilkan dengan format pemisah ribuan untuk mempermudah pembacaan data.

- Program ini menganalisis data pembelian berdasarkan kota dan kategori produk. Langkah pertama adalah memeriksa apakah kolom yang diperlukan, yaitu "customer_city", "order_item_id", dan "product_category_name", ada dalam dataset. Selanjutnya, data dikelompokkan berdasarkan kota untuk menghitung jumlah pembelian total (dihitung dengan menghitung jumlah `order_item_id`), yang kemudian diurutkan berdasarkan jumlah pembelian terbanyak. Dari data yang sudah terurut, program memilih 10 kota dengan jumlah pembelian tertinggi. Data ini kemudian disaring untuk hanya mencakup kota-kota teratas tersebut. Setelah itu, program mengelompokkan data berdasarkan kota dan kategori produk untuk menghitung jumlah pembelian per kategori produk di setiap kota. Kemudian, program menyaring produk dengan jumlah pembelian terbanyak di setiap kota menggunakan fungsi `nlargest`. Hasil akhirnya adalah menampilkan kota-kota teratas beserta kategori produk yang memiliki jumlah pembelian tertinggi di masing-masing kota, dengan format pemisah ribuan pada jumlah pembelian agar lebih mudah dibaca.

```python
import pandas as pd

# Memeriksa apakah kolom yang diperlukan ada
required_columns = ["customer_city", "order_item_id", "product_category_name"]
if not all(col in Product_stok_kota_final.columns for col in required_columns):
    raise ValueError(f"Data harus memiliki kolom: {required_columns}")

# Mengelompokkan data berdasarkan kota dan menghitung jumlah total pembelian (order_item_id)
grouped_data = Product_stok_kota_final.groupby("customer_city").agg(
    total_purchases=("order_item_id", "count")  # Menghitung jumlah pembelian per kota
).reset_index()

# Mengurutkan data berdasarkan jumlah pembelian terbanyak
sorted_data = grouped_data.sort_values(by="total_purchases", ascending=False)

# Menampilkan 10 kota dengan jumlah pembelian terbanyak
top_10_cities = sorted_data.head(10)

# Menyaring data hanya untuk 10 kota teratas
top_10_cities_data = Product_stok_kota_final[Product_stok_kota_final['customer_city'].isin(top_10_cities['customer_city'])]

# Mengelompokkan data berdasarkan kota dan kategori produk, dan menghitung jumlah pembelian per kategori produk
grouped_products = top_10_cities_data.groupby(["customer_city", "product_category_name"]).agg(
    total_product_sales=("order_item_id", "count")  # Menghitung jumlah pembelian per kategori produk di setiap kota
).reset_index()

# Menyaring produk dengan pembelian terbanyak di setiap kota menggunakan nlargest
top_products_per_city = grouped_products.groupby("customer_city").apply(
    lambda x: x.nlargest(1, 'total_product_sales')
).reset_index(drop=True)

# Menampilkan hasil
top_products_per_city[['customer_city', 'product_category_name', 'total_product_sales']].style.format({
    'total_product_sales': '{:,}'  # Menampilkan jumlah pembelian dengan pemisah ribuan
})
```

### Pertanyaan 1:

Program ini dimulai dengan memverifikasi apakah data yang digunakan memiliki kolom yang diperlukan, yaitu product_category_name, review_score, dan order_item_id. Setelah memastikan keberadaan kolom tersebut, data kemudian dikelompokkan berdasarkan kategori produk (product_category_name) untuk menghitung total pembelian (purchase_count) dan rata-rata rating (average_rating). Data yang telah dikelompokkan ini kemudian disortir untuk mendapatkan produk dengan pembelian terbanyak dan rating tertinggi. Selanjutnya, program menggabungkan dua hasil sorting tersebut untuk mendapatkan produk yang memiliki pembelian lebih dari 5000 dan rating lebih dari 4. Program akhirnya menampilkan lima produk teratas yang memenuhi kriteria tersebut dalam bentuk visualisasi bar chart interaktif menggunakan Plotly, yang memungkinkan pengguna untuk melihat total pembelian dan rating produk dengan lebih jelas.

```python
import pandas as pd
import plotly.express as px

# Memeriksa apakah kolom yang diperlukan ada
required_columns = ["product_category_name", "review_score", "order_item_id"]
if not all(col in product_favorit_akhir.columns for col in required_columns):
    raise ValueError(f"Data harus memiliki kolom: {required_columns}")

# Mengelompokkan data berdasarkan kategori produk dan menghitung pembelian unik berdasarkan order_item_id
grouped_data = product_favorit_akhir.groupby("product_category_name").agg(
    purchase_count=("order_item_id", "sum"),  # Menghitung jumlah pembelian
    average_rating=("review_score", "mean")  # Menghitung rata-rata rating
).reset_index()

# Menyaring produk dengan pembelian terbanyak dan rating tertinggi
top_purchased_products = grouped_data.sort_values(by="purchase_count", ascending=False)
top_rated_products = grouped_data.sort_values(by="average_rating", ascending=False)

# Menampilkan produk yang ada di kedua kategori: pembelian terbanyak dan rating tertinggi
top_combined_products = pd.merge(top_purchased_products[['product_category_name', 'purchase_count']],
                                  top_rated_products[['product_category_name', 'average_rating']],
                                  on="product_category_name")

# Menyaring produk dengan purchase_count > 5000 dan average_rating > 4
filtered_products = top_combined_products[(top_combined_products['purchase_count'] > 5000) &
                                           (top_combined_products['average_rating'] > 4)]

# Menampilkan 5 produk dengan purchase_count tertinggi yang memenuhi kriteria
top_5_purchased_products = filtered_products.head(5)

# Membuat visualisasi interaktif menggunakan Plotly
fig = px.bar(top_5_purchased_products,
             x='purchase_count',
             y='product_category_name',
             text='purchase_count',  # Menambahkan nilai pada bar
             hover_data={'purchase_count': True, 'average_rating': True},  # Menambahkan data saat hover
             labels={'purchase_count': 'Purchase Count', 'product_category_name': 'Product Category'},
             title="Top 5 Products with Highest Purchase Count and Rating Above 4")

# Menampilkan grafik dengan interaktif
fig.update_traces(texttemplate='%{text}', textposition='inside', hoverinfo="x+y+text")
fig.show()
```

Produk dengan pembelian terbanyak di setiap kota dipilih, dan akhirnya, visualisasi interaktif dibuat menggunakan Plotly, menampilkan total penjualan produk di setiap kota dengan informasi tambahan saat hover. Hasilnya adalah grafik batang yang memperlihatkan produk terlaris per kota.

```python
import pandas as pd
import plotly.express as px

# Memeriksa apakah kolom yang diperlukan ada
required_columns = ["customer_city", "order_item_id", "product_category_name"]
if not all(col in Product_stok_kota_final.columns for col in required_columns):
    raise ValueError(f"Data harus memiliki kolom: {required_columns}")

# Mengelompokkan data berdasarkan kota dan menghitung jumlah total pembelian (order_item_id)
grouped_data = Product_stok_kota_final.groupby("customer_city").agg(
    total_purchases=("order_item_id", "count")  # Menghitung jumlah pembelian per kota
).reset_index()

# Mengurutkan data berdasarkan jumlah pembelian terbanyak
sorted_data = grouped_data.sort_values(by="total_purchases", ascending=False)

# Menampilkan 10 kota dengan jumlah pembelian terbanyak
top_10_cities = sorted_data.head(10)

# Menyaring data hanya untuk 10 kota teratas
top_10_cities_data = Product_stok_kota_final[Product_stok_kota_final['customer_city'].isin(top_10_cities['customer_city'])]

# Mengelompokkan data berdasarkan kota dan kategori produk, dan menghitung jumlah pembelian per kategori produk
grouped_products = top_10_cities_data.groupby(["customer_city", "product_category_name"]).agg(
    total_product_sales=("order_item_id", "count")  # Menghitung jumlah pembelian per kategori produk di setiap kota
).reset_index()

# Menyaring produk dengan pembelian terbanyak di setiap kota menggunakan nlargest
top_products_per_city = grouped_products.groupby("customer_city").apply(
    lambda x: x.nlargest(1, 'total_product_sales')
).reset_index(drop=True)

# Membuat diagram interaktif dengan Plotly
fig = px.bar(
    top_products_per_city,
    x='total_product_sales',
    y='customer_city',
    color='product_category_name',
    title='Top Products by City',
    labels={'total_product_sales': 'Total Product Sales', 'customer_city': 'Customer City', 'product_category_name': 'Product Category'},
    text='total_product_sales',  # Menambahkan nilai total pembelian sebagai teks di atas batang
    hover_data={'customer_city': True, 'product_category_name': True, 'total_product_sales': True}  # Menampilkan data saat hover
)

# Menampilkan plot interaktif
fig.update_traces(textposition='outside', texttemplate='%{text}', hoverinfo='x+y+text+name')

fig.update_layout(
    xaxis_title="Total Product Sales",
    yaxis_title="Customer City",
    template="plotly_dark"
)

fig.show()
```
